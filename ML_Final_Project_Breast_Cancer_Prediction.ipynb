```python
X= extracted_data.drop('Class', axis=1)
y= extracted_data['Class']
```

SPLITTING THE DATA

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler

X_train, X_test, y_train,y_test= train_test_split(X,y, test_size=0.3)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)


sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Model 1 is DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.tree import DecisionTreeClassifier

DTC = DecisionTreeClassifier(random_state=42).fit(X_train,y_train)
y1_pred = DTC.predict(X_test)

score_DTC = accuracy_score(y_test,y1_pred)*100

print("The accuracy of DecisionTreeClassifier model is", score_DTC)
print("MSE is: ", mean_squared_error(y_test,y1_pred))
print("R2 SCORE: ",r2_score(y_test,y1_pred))
```

```python
#Model 2 is Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
RFC= RandomForestClassifier(random_state=42,verbose=False).fit(X_train,y_train)
y2_pred=RFC.predict(X_test)

score_RFC = accuracy_score(y_test,y2_pred)*100

print("The accuracy of RandomForestClassifier model is", score_RFC)
print(mean_squared_error(y_test,y2_pred))
print(r2_score(y_test,y2_pred))
```

```python
#Model 3 is Deep Neural Network

import tensorflow as tf
from tensorflow import keras

model = keras.models.Sequential([
keras.layers.Flatten(input_shape=[10,]),
keras.layers.Dense(180, activation="elu"),
keras.layers.Dense(150, activation="relu"),
keras.layers.Dense(60, activation="relu"),
keras.layers.Dense(30, activation="softmax"),
keras.layers.Dense(1),
keras.layers.Dropout(rate=0.5),
])

model.summary()
```

```python
model.compile(loss='BinaryCrossentropy', metrics=['accuracy'])
history = model.fit(X_train,y_train, epochs=100, validation_data=(X_test, y_test), verbose=2)
```

```python
predictions = model.predict(X_test)

y_DNN = (predictions > 0.5)

score_DNN = accuracy_score(y_test, y_DNN)*100

print('The accuracy of DNN Model',score_DNN)
```

```python
#Accuracy Comparision.
import matplotlib.pyplot as plt
a=score_DTC
b=score_RFC
c=score_DNN

height = [a,b,c]
bars = ('Decision Tree Classifier','Random Forest Classifier','DNN')
y_pos = np.arange(len(bars))

# Create bars
plt.bar(y_pos, height)

# Create names on the x-axis
plt.xticks(y_pos, bars)

# Show graphic
plt.show()
```

```python
#generating confusion matrix
from sklearn.metrics import confusion_matrix
def plot_confusion_matrix(y, y2_pred):
    acc = round(accuracy_score(y, y2_pred), 2)
    cm = confusion_matrix(y, y2_pred)
    sns.heatmap(cm, annot=True)
    plt.xlabel('Predections',size=15)
    plt.ylabel('Test set',size=15)
    plt.title('Accuracy Score: {:.0%}'.format(acc), size=15)
    plt.show()
plot_confusion_matrix(y_test, y2_pred)
```
